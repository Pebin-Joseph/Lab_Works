```python
import random
import time
import matplotlib.pyplot as plt
import numpy as np

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def measure_search_time(arr, target, num_trials=100):
    total_time = 0

    for _ in range(num_trials):
        start_time = time.perf_counter()
        linear_search(arr, target)
        end_time = time.perf_counter()
        total_time += (end_time - start_time)

    return (total_time / num_trials) * 1000

def analyze_linear_search(n=10000):
    print(f"Starting linear search analysis with {n} elements...")
    print("=" * 60)

    print("Creating random array...")
    arr = [random.randint(1, 100000) for _ in range(n)]

    best_case_target = arr[0]
    middle_index = n // 2
    average_case_target = arr[middle_index]
    worst_case_target = max(arr) + 1

    print(f"Array ready! Size: {n}")
    print(f"Best case: Looking for {best_case_target} (it's at position 0)")
    print(f"Average case: Looking for {average_case_target} (it's at position {middle_index})")
    print(f"Worst case: Looking for {worst_case_target} (it's not in the array)")
    print()

    print("Running timing tests...")
    best_time = measure_search_time(arr, best_case_target)
    average_time = measure_search_time(arr, average_case_target)
    worst_time = measure_search_time(arr, worst_case_target)

    print("Results are in!")
    print(f"Best case took: {best_time:.6f} ms")
    print(f"Average case took: {average_time:.6f} ms")
    print(f"Worst case took: {worst_time:.6f} ms")
    print()

    print("Why these results make sense:")
    print("Best Case - Lightning fast!")
    print("  Found the element immediately at position 0")
    print("  No matter how big the array gets, this stays constant")
    print()

    print("Average Case - Pretty decent")
    print("  Had to check about half the elements")
    print("  Gets slower as the array grows")
    print()

    print("Worst Case - Had to work hard")
    print("  Checked every single element and still didn't find it")
    print("  This is why big datasets can be painful")
    print()

    return {
        'best_time': best_time,
        'average_time': average_time,
        'worst_time': worst_time,
        'size': n
    }

def plot_comparison(results):
    cases = ['Best Case\n(Found instantly)', 'Average Case\n(Found halfway)', 'Worst Case\n(Not found)']
    times = [results['best_time'], results['average_time'], results['worst_time']]
    colors = ['lightgreen', 'gold', 'lightcoral']

    plt.figure(figsize=(12, 7))
    bars = plt.bar(cases, times, color=colors, alpha=0.8, edgecolor='black', linewidth=1.5)

    for bar, time in zip(bars, times):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.001,
                f'{time:.6f} ms', ha='center', va='bottom', fontweight='bold', fontsize=11)

    plt.title(f'How Fast is Linear Search? (Testing {results["size"]} elements)',
              fontsize=16, fontweight='bold', pad=20)
    plt.ylabel('Time (milliseconds)', fontsize=13)
    plt.xlabel('Different Scenarios', fontsize=13)
    plt.grid(axis='y', alpha=0.3, linestyle='--')
    plt.tight_layout()
    plt.show()

def analyze_input_size_impact():
    sizes = [1000, 5000, 10000, 25000, 50000]
    best_times = []
    average_times = []
    worst_times = []

    print("Let's see how array size affects performance...")
    print("=" * 60)

    for size in sizes:
        print(f"Testing with {size} elements...")
        result = analyze_linear_search(size)
        best_times.append(result['best_time'])
        average_times.append(result['average_time'])
        worst_times.append(result['worst_time'])
        print(f"Done with {size} elements!\n")

    plt.figure(figsize=(14, 10))

    plt.subplot(2, 1, 1)
    plt.plot(sizes, best_times, 'g-o', label='Best Case (Super fast!)', linewidth=3, markersize=8)
    plt.plot(sizes, average_times, 'orange', marker='s', label='Average Case (Pretty good)', linewidth=3, markersize=8)
    plt.plot(sizes, worst_times, 'r-^', label='Worst Case (Slow but sure)', linewidth=3, markersize=8)

    plt.title('How Array Size Affects Search Speed', fontsize=16, fontweight='bold')
    plt.xlabel('Number of Elements', fontsize=13)
    plt.ylabel('Time (milliseconds)', fontsize=13)
    plt.legend(fontsize=12)
    plt.grid(True, alpha=0.3)

    plt.subplot(2, 1, 2)
    plt.loglog(sizes, best_times, 'g-o', label='Best Case (Stays flat)', linewidth=3, markersize=8)
    plt.loglog(sizes, average_times, 'orange', marker='s', label='Average Case (Goes up)', linewidth=3, markersize=8)
    plt.loglog(sizes, worst_times, 'r-^', label='Worst Case (Goes up faster)', linewidth=3, markersize=8)

    plt.title('The Same Data on Log Scale (Notice the Patterns)', fontsize=16, fontweight='bold')
    plt.xlabel('Number of Elements', fontsize=13)
    plt.ylabel('Time (milliseconds)', fontsize=13)
    plt.legend(fontsize=12)
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    print("Here's the full breakdown:")
    print("-" * 80)
    print(f"{'Elements':<12} {'Best (ms)':<15} {'Average (ms)':<15} {'Worst (ms)':<15} {'Ratio':<10}")
    print("-" * 80)
    for i, size in enumerate(sizes):
        ratio = worst_times[i] / best_times[i]
        print(f"{size:<12} {best_times[i]:<15.6f} {average_times[i]:<15.6f} {worst_times[i]:<15.6f} {ratio:<10.1f}x")
    print("-" * 80)

def demonstrate_linear_search():
    print("Let's see linear search in action!")
    print("=" * 40)

    demo_arr = [64, 34, 25, 12, 22, 11, 90]
    print(f"Our test array: {demo_arr}")
    print()

    targets = [64, 22, 99]
    scenarios = ["first element", "middle element", "element not in array"]

    for target, scenario in zip(targets, scenarios):
        print(f"Looking for {target} ({scenario})...")
        start_time = time.perf_counter()
        result = linear_search(demo_arr, target)
        end_time = time.perf_counter()

        if result != -1:
            print(f"Found it! {target} is at position {result}")
        else:
            print(f"Nope, {target} isn't here")
        print(f"That took {(end_time - start_time) * 1000:.6f} ms")
        print()

if __name__ == "__main__":
    print("Welcome to Linear Search Analysis!")
    print("We're going to test how fast we can find stuff in arrays")
    print()

    demonstrate_linear_search()

    print("Now let's do some serious testing...")
    result = analyze_linear_search(10000)
    plot_comparison(result)

    analyze_input_size_impact()

    print("\nWhat we learned:")
    print("=" * 60)
    print(" Best case is always fast - doesn't matter how big the array is")
    print(" Average case gets slower as arrays get bigger")
    print(" Worst case is the real killer - scales directly with size")
    print(" The bigger the array, the bigger the difference between best and worst")
    print(" For huge datasets, you really want to avoid worst-case scenarios")
    print(" This is why other search algorithms (like binary search) exist!")
    print()
    print("Thanks for exploring linear search with us! ")
```
